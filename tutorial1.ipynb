```python

# --- Task 3 ----

# printing dataset to check content
print(dataset)

# splitting data into 4 variables
X_train, X_test, Y_train, Y_test = train_test_split(features, label, train_size=0.6)
print(X_train, X_test, Y_train, Y_test)

# importing algorithm
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()

# training the data
tree.fit(X_train, Y_train)
pred_sample = tree.predict([[13.17, 2.59, 2.37, 20, 120, 1.65, .68, .53, 1.46, 9.3, .6, 1.62, 840]])
print(pred_sample)

import numpy as np

# testing sample in X_test
y_pred = tree.predict(X_test)
print("Expected >>>", np.array(Y_test))
print("Prediction >", y_pred)
```

```python
# --- Task 4 ---

# importing algorithm
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier()

# splitting data into 4 variables
X_train, X_test, Y_train, Y_test = train_test_split(features, label, train_size=0.6)
print(X_train, X_test, Y_train, Y_test)

# training the algorithm
d_tree.fit(X_train, Y_train)
pred_sample = d_tree.predict([[13.17, 2.59, 2.37, 20, 120, 1.65, .68, .53, 1.46, 9.3, .6, 1.62, 840]])
print(pred_sample)

import numpy as np

# testing sample in X_test
y_pred = d_tree.predict(X_test)
print("Expected >>>", np.array(Y_test))
print("Prediction >", y_pred)

# checking accuracy of model

#using sklearn
from sklearn.metrics import accuracy_score
y_true = [1,0,0,1,1]
y_pred = [1,1,1,1,1]
acc = accuracy_score(y_true, y_pred)
print("Accuracy=", 100*acc)

#using matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.figure()
# plt.plot([1,2,3,4], [1,4,2,3])
# plt.scatter([1,2,3,4], [1,4,2,3])
# plt.bar([1,2,3,4], [1,4,2,3])
plt.hist2d([1,2,3,4], [1,4,2,3])
plt.xlabel("Class ")
plt.ylabel("Alcohol")
plt.show()
```
