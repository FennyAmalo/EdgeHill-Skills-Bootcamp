```python
# Exercise 4a: Sentiment Analysis

import nltk
nltk.download('punkt')

# analysing the sentiment - using textblob library

import pandas as pd
df = pd.read_csv('Video_games_reviews.csv', delimiter = '\t', header = None)
from textblob import TextBlob

'''
Iterate through video reviews, and for each review, call textblob
to analyse the sentiment of that review
'''

for index, review_text in enumerate(video_review_texts):
  # call textblob for current review text and print the sentiment scores of each sentence
  blob = TextBlob(review_text)
  print('Analysing review\t' + review_text)
  # iterate through the sentence of current review_text and print the sentiment scores of each sentence
  for sentence in blob.sentences:
    print('------------SENTIMENT OF SENTENCE-----------')
    print(sentence, '\t', sentence.sentiment.polarity)
    print('------------END--------------')




```

```python
# NTLK libraries for NER

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_de_chunker')
nltk.download('words')

from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree

input_sentence = 'I study at Edge Hill University, which is located in Ormskirk'

# perform tokenisation, pos tagging and ner in only one line of code
chunks = ne_chunk(pos_tag(word_tokenize(input_sentence)))
print(chunks)

```

```python
# iterate through chunks
for ne in chunks:
   #if current chunk is a named entity type, then print it
   if type(ne) == Tree:
     print(ne)
```

```python
from nltk.tokenize import sent_tokenize
# Exercise 6: Apply NER to video game dataset

df = pd.read_csv('Video_games_reviews.csv', delimiter = '\t', header = None)
video_review_texts = df[2]

for index, review_text in enumerate(video_review_texts):
  for sentence in sent_tokenize(review_text):
    print('Sentence:\t', sentence)

    # perform tokenisation, pos tagging and ner in only one line of code
    chunks = ne_chunk(pos_tag(word_tokenize(sentence)))

    # iterate through chunks
    for ne in chunks:
      # if current chunk is a named entity type then print it
      if type(ne) == Tree:
        print('Named Entities:\t', ne)

    print('')
```
